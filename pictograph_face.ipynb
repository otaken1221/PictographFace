```python
import face_recognition
from PIL import Image, ImageDraw
import numpy as np
```

```python
im = Image.open('ai_model_09.png')
img_pien = Image.open('pien-removebg-preview.png')

image = np.asarray(im)
face_bboxes = face_recognition.face_locations(image)

img_clear = Image.new("RGBA", im.size, (255, 255, 255, 0))
img_clear.paste(im, (0,0))

img_pien_resize = img_pien.resize((face_bboxes[0][2]-face_bboxes[0][0], face_bboxes[0][1]-face_bboxes[0][3]))

img_clear.paste(img_pien_resize, (face_bboxes[0][3], face_bboxes[0][0]))
image = np.asarray(img_clear)
```

```python
# 表示用の処理
# Pillowというライブラリを使用して、PIL形式のイメージを作成
pil_image = Image.fromarray(image)
# PillowのImageDrawインスタンスを作成
draw = ImageDraw.Draw(pil_image)


# 結果の表示
display(pil_image)
```
